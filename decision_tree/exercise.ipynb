```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df.drop(columns=['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df.isnull().sum()
```

```python
inputs = df.drop(columns=['Survived'])
target = df['Survived']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
pclass_le = LabelEncoder()
sex_le = LabelEncoder()
inputs['Pclass_n'] = pclass_le.fit_transform(inputs['Pclass'])
inputs['Sex_n'] = sex_le.fit_transform(inputs['Sex'])
```

```python
inputs = inputs.drop(columns=['Sex', 'Age'])
inputs
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.1)
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict(X_test)
```
